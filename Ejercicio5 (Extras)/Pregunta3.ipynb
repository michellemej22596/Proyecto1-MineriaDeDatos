#### Pregunta Extra 3. ¿Las peliculas con mayor duracion tienen mayores ingresos?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv('../movies.csv', encoding='latin1')

#hacer el diagrama de dispersion
plt.scatter(df['runtime'], df['revenue'], color='green', alpha=0.6)
plt.title('Relación entre duracion e ingresos')
plt.xlabel('Duracion (runtime)')
plt.ylabel('Ingreso (revenue)')
plt.grid(alpha=0.5)
plt.show()

#ver el coeficiente de correlacion entre ambos datos 
correlacion = df["runtime"].corr(df["revenue"])
print("El coeficiente de correlacion entre la duracion y los ingrersos de la pelicula es", correlacion)
```
