#### Inciso K. ¿Cómo se correlacionan los presupuestos con los ingresos? ¿Los altos presupuestos significan altos ingresos? Haga los gráficos que necesite, histograma, diagrama de dispersión 

Aqui quiero comparar lo que costo hacer la pelicula con lo que gano, para determinar si las peliculas que costaron mas de hacer son las que mas ganan, o cual es la relacion entre ambos datos, si existe alguna.

Pensamos que algo que nos puede ayudar a sacar esa conclusion es graficar en el eje x la variable de presupuesto y en el eye y la variable de ingresos, y ver si hay una tendencia. 
Ademas, calculamos el coeficiente de correlacion para ver un dato numerico que nos indique si la estadistica dice que estan correlacionados o no

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar los datos
df = pd.read_csv('../movies.csv', encoding='latin1')

#hacer el diagrama de dispersion
plt.scatter(df['budget'], df['revenue'], color='brown', alpha=0.6)
plt.title('Relación entre presupuesto e ingresos')
plt.xlabel('Presupuesto (budget)')
plt.ylabel('Ingreso (revenue)')
plt.grid(alpha=0.5)
plt.show()

#ver el coeficiente de correlacion entre ambos datos 
correlacion = df["budget"].corr(df["revenue"])
print("El coeficiente de correlacion entre la variable de presupueso y la de ingrersos es", correlacion)
```
