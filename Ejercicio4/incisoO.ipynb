#### Inciso O ¿Qué estrategias de marketing, como videos promocionales o páginas oficiales, generan mejores resultados?

variables para marketing: video (si,no).
variables para mejores resultados: revenue, popularity, voteAvg,

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
df = pd.read_csv('../movies.csv', encoding='latin1')

# Convertir las columnas necesarias a valores numéricos
df["popularity"] = pd.to_numeric(df["popularity"], errors="coerce")

# Crear las columnas binarias para las estrategias de marketing
df["hasVideo"] = df["video"].astype(str).str.upper() == "TRUE"  # True si es "TRUE"
df["hasHomePage"] = df["homePage"].notna()  # True si tiene página oficial

# Validar que existan valores True y False
print(df["hasVideo"].value_counts())
print(df["hasHomePage"].value_counts())

# Gráfica 1: Impacto de Videos Promocionales en Popularidad
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["hasVideo"], y=df["popularity"])
plt.title("Impacto de Videos Promocionales en Popularidad")
plt.xlabel("¿Tiene Video Promocional?")
plt.ylabel("Popularidad (Popularity)")
plt.show()

# Gráfica 2: Impacto de Páginas Oficiales en Popularidad
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["hasHomePage"], y=df["popularity"])
plt.title("Impacto de Páginas Oficiales en Popularidad")
plt.xlabel("¿Tiene Página Oficial?")
plt.ylabel("Popularidad (Popularity)")
plt.show()
```
