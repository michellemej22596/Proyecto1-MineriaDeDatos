#### Inciso L. ¿Se asocian ciertos meses de lanzamiento con mejores ingresos? 

Quiero ver si los mejores lanzamientos han sido en meses especificos. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv('../movies.csv', encoding='latin1')

# Extraer el mes de la fecha de lanzamiento
df['releaseDate'] = pd.to_datetime(df['releaseDate'], errors='coerce')
df['releaseMonth'] = df['releaseDate'].dt.month

# Calcular los ingresos promedio por mes
ingresomensual= df.groupby('releaseMonth')['revenue'].mean()

# Visualización de ingresos por mes
plt.figure(figsize=(10, 6))
sns.barplot(x=ingresomensual.index, y=ingresomensual.values, palette="coolwarm")
plt.xlabel("Mes de lanzamiento")
plt.ylabel("Ingresos promedio ($)")
plt.title("Ingresos promedio por mes de lanzamiento")
plt.xticks(range(12), ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"])
plt.show()
```

```python
# Cargar los datos
df = pd.read_csv('../movies.csv', encoding='latin1')

# Convertir la columna de fechas a formato datetime
df["releaseDate"] = pd.to_datetime(df["releaseDate"], errors="coerce")

# Filtrar las 1000 películas con mayores ingresos
top1000ingresos= df.sort_values(by="revenue", ascending=False).head(1000)

# Extraer el mes de lanzamiento
top1000ingresos["releaseMonth"] = top1000ingresos["releaseDate"].dt.month

# Contar la frecuencia de cada mes
monthly_counts = top1000ingresos["releaseMonth"].value_counts().sort_index()

# Crear una tabla con los meses como filas y la frecuencia como valores
monthly_table = pd.DataFrame({"Frecuencia": monthly_counts})

# Mostrar la tabla
print(monthly_table)
```
