#### Inciso N. ¿Cómo se correlacionan las calificaciones con el éxito comercial?

variables interesantes.
para exito comercial: popularity, revenue
para calificaciones: voteAvg

planeo hacer un graifco de revenue vs voteAvg y de popularity vs VoteAvg y sus respectivas correlaciones

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar los datos
df = pd.read_csv('../movies.csv', encoding='latin1')

#hacer el diagrama de dispersion
plt.scatter(df['voteAvg'], df['revenue'], color='brown', alpha=0.6)
plt.title('Relación entre ingresos y calificacion promedio')
plt.xlabel('Calificacion Promedio (voteAvg)')
plt.ylabel('Ingreso  (revenue)')
plt.grid(alpha=0.5)
plt.show()

#ver el coeficiente de correlacion entre ambos datos 
correlacion = df["voteAvg"].corr(df["revenue"])
print("El coeficiente de correlacion entre la variable de ingreso y la calificacion promedio es", correlacion)
```

```python
# Cargar los datos
df = pd.read_csv('../movies.csv', encoding='latin1')

#hacer el diagrama de dispersion
plt.scatter(df['voteAvg'], df['popularity'], color='brown', alpha=0.6)
plt.title('Relación entre popularidad y calificacion promedio')
plt.xlabel('Calificacion Promedio (voteAvg)')
plt.ylabel('Popularidad  (popularity)')
plt.grid(alpha=0.5)
plt.show()

#ver el coeficiente de correlacion entre ambos datos 
correlacion = df["voteAvg"].corr(df["popularity"])
print("El coeficiente de correlacion entre la variable de popularidad y la calificacion promedio es", correlacion)
```
